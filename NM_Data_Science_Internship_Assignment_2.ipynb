**Task 2: Data Preprocessing**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
z = data[["Selling_Price"]]
x = scaler.fit_transform(z)
data[["Selling_Price"]] = x
```

```python
data[["Present_Price"]] = data[["Present_Price"]]
z = data[["Present_Price"]]
x = scaler.fit_transform(z)
data[["Present_Price"]] = x
```

```python
y = data[["Year"]]
s = scaler.fit_transform(y)
data[["Year"]] = s
```

```python
data.drop('Car_Name', axis=1)
```

**Task 3: Perform train_test_split**

```python
from sklearn.model_selection import train_test_split
x = data.drop('Car_Name', axis=1)
y = data["Car_Name"]
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.25, random_state = 20)
```

Thus, we have split the data as 75% training data and remaining 25% as test data
