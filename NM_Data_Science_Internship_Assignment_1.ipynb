**Assignment 1**

Purpose: To import the libraries and load the Titanic dataset

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset("titanic")
```

```python
df.head()
```

**Q1)What is the overall survival rate of passengers on the Titanic?**

```python
survived = pd.Series(df['alive'].value_counts())
print(survived)
plt.pie(survived, labels=['dead','alive'], autopct="%.2f")
plt.legend()
plt.title("Survival rate of passengers")
plt.show()
```

**Q2)What was the gender distribution among the passengers on the Titanic?**

```python
gender = pd.Series(df['sex'].value_counts())
print(gender)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("No. of Passengers")
sns.barplot(x=gender.index,y=gender.values,data=df)
```

**Q3)Did the survival rate differ by gender?If so, how much?**

```python
gender_surv = df.groupby('sex')['alive'].value_counts()
print(gender_surv)
plt.title("Distribution of Survival Rate Based on Gender")
sns.barplot(x='sex',y='survived',data=df)
```

**Q4)What was the age distribution among the passengers on the Titanic?**

```python
age_dist = pd.Series(df['age'].value_counts())
print(age_dist)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("No. of Passengers")
sns.barplot(x=age_dist.index,y=age_dist.values,data=df)
```

**Q5)Did the survival rate differ by the journey class?If so, how much?**

```python
pclass_surv = df.groupby('pclass')['alive'].value_counts()
print(pclass_surv)
plt.title("Distribution of Survival Rate Based on Journey Class")
sns.barplot(x='pclass',y='survived',data=df)
```
